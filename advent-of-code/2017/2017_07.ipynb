```python
data = aocd.get_data("session_id",7,2017)
```

```python
data = [line.replace(' -> ', ' ').split() for line in data.split('\n')]
```

```python
#part 1
def get_parents_children(data):
    children = set()
    parents = set()
    for line in data:
        parents|= set([line[0]])
        children |= set(map(lambda x: x.rstrip(','), line[2:]))
        
    return children, parents

children, parents = get_parents_children(data)

def get_bottom(parents, children):
    for root in parents:
        if root not in children:
            return root

bottom = get_bottom(parents, children)
```

```python
# part 2
def get_program_dict(data):
    program = {}
    for line in data:
        program[line[0]] = [int(line[1].lstrip('(').rstrip(')')), map(lambda x: x.rstrip(','), line[2:]), 0]
    
    return program

program = get_program_dict(data)

def compute_weights(program, root):
    if len(program[root][1]) == 0:
        program[root][2] = program[root][0]
        return program[root][2]
    
    else:
        branch_sum = 0
        for child in program[root][1]:
            branch_sum += compute_weights(program, child)
        program[root][2] = program[root][0] + branch_sum
        return program[root][2]
    
program[bottom][2] = compute_weights(program, bottom)    
```
